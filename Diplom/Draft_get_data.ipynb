```python
import time
from bs4 import BeautifulSoup
import sys, io
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.proxy import *
```

```python
driver = webdriver.Safari()
wait = WebDriverWait( driver, 10 )


# Append your app store urls here
urls = ['https://play.google.com/store/apps/details?id=com.dc.hwsj']

for url in urls:

    driver.get(url)
    page = driver.page_source

    soup_expatistan = BeautifulSoup(page, "html.parser")

    #expatistan_table = soup_expatistan.find("div", class_="id-app-title")

    #print("App name: ", expatistan_table.string)

    #expatistan_table = soup_expatistan.find("div", itemprop="numDownloads")

    #print("Installs Range: ", expatistan_table.string)

    #expatistan_table = soup_expatistan.find("meta", itemprop="ratingValue")

    #print("Rating Value: ", expatistan_table['content'])

    app_info = {'reviews': {'comments': [] }}
 
    reviews_div = driver.find_element_by_xpath('//span[contains(text(), "reviews"]').get_attribute("innerHTML")
    soup_expatistan = BeautifulSoup(reviews_div, "html.parser")
    
    
    for review in soup_expatistan.findAll("div", class_="single-review"):
                author = review.find('span', {'class': 'author-name'}).text.rstrip().lstrip()
                authorId = find_between(str(review.find('span', {'class': 'author-name'})), '?id=', '"')
                date = review.find('span', {'class': 'review-date'}).text
                message = review.find('div', {'class': 'review-body'}).text.replace('Resenha completa', '').rstrip().lstrip()
                app_info['reviews']['comments'].append({'author': author, 'authorId': authorId, 'date': date, 'message': message.encode('utf-8')})
                print(message)
            
driver.quit()
```
